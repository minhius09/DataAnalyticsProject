First we look for the columns needed to be done some folowings: 
* Drop any columns that have the repeated values because they are useless for further analysis

Now take a closer look for each individual columns 

```python
autos.loc[:10,:]
```

I noticed that 'price' and 'odometer' columns having numeric data but stored as text because '$', 'km', ','. So next I will convert them to numeric values 

* First I will eliminate '$', 'km' and ',' 

```python
autos['price'] = autos['price'].str.replace('$','').str.replace(',','')
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',','')
autos[['price','odometer']].head()
```

* Because I have removed '$','km' and ','. Next step, I will convert them into int type

```python
autos[['price','odometer']]=autos[['price','odometer']].astype(int)
```

* Last step I need to rename columns for not missing any information about monetary unit and measurment unit 

```python
autos.rename({'price' : 'price_$','odometer':'odometer_km'}, axis = 1, inplace = True)
```

I need more investigation about these columns: 'seller' , 'offer_type', 'unrepaired_damage', 'nr_of_pictures' to check if they have the repeated value or not

There are 3 columns are useless: 'seller' , 'offer_type' and 'nr_of_pictures' because they have mostly one value. Next I will remove them 

```python
#remove columns
autos = autos.drop(['seller','offer_type','nr_of_pictures'],axis = 1)
```

#### Analyzing prices by brand

```python
autos['brand'].value_counts(normalize = True).map('{:.2%}'.format)
```

Next I want to explore the mean price of the top 20 brand 

```python
the_20_brand = autos['brand'].value_counts(normalize = True).map('{:.2%}'.format).head(20)
```

```python
list_20_brand = the_20_brand.index
list_20_brand
```

I will create a empty dictionary for using loop technique 

```python
mean_price_by_brand = {}
```

```python
for brand in list_20_brand:
    price = autos.loc[autos['brand'] == brand, 'price_$']
    mean_price_by_brand[brand] = price.mean()
mean_price_by_brand
```

I will order the dictionary by descending mean price 

```python
sorted(mean_price_by_brand.items(), key= lambda x: x[1], reverse = True)
```

According to the above analysis:

'sonstige_autos' is the most expensive car and 'renault' is the least expensive car in our selected top German brands.

```python
mean_mileage_by_brand = {}
```

```python
for brand in list_20_brand:
    mileage = autos.loc[autos['brand'] == brand, 'odometer_km']
    mean_mileage_by_brand[brand] = mileage.mean()
mean_mileage_by_brand
```

```python
pd.DataFrame({'mean_price':pd.Series(mean_price_by_brand),'mean_mileage':pd.Series(mean_mileage_by_brand)})
```

There is also a relationship between mean price and mean mileage. Greater price means lesser mileage
